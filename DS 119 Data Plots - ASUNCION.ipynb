As a startup, we use Matplotlib since it is perhaps the most “standard” Python plotting library.

## Creating Plots with Matplotlib

```python
# this line tells Jupyter to display matplotlib figures in the notebook
%matplotlib inline

import matplotlib.pyplot as plt

# create a bivariate dataset
data_x = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]
data_y = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
```

```python
# start setting up the plotting area
fig, ax = plt.subplots()

ax.scatter(x=data_x, y=data_y, c="#E69F00")
```

```python
fig, ax = plt.subplots()

ax.scatter(x=data_x, y=data_y, c="#E69F00")

# for labelling axes and the entire figure, you may use the following commands

ax.set_xlabel("data_x")
ax.set_ylabel("data_y")
ax.set_title("Sample Data Plot")
```

```python
data2_y = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
data2_y_scaled = [y*2.0 for y in data2_y]

fig, ax = plt.subplots()

ax.scatter(x=data_x, y=data_y, c="Blue", label = "set 1")
ax.scatter(x=data_x, y=data2_y, c="Red", label = "set 2")
ax.scatter(x=data_x, y=data2_y_scaled, c="Green", label = "set 2 (scaled)")

ax.set_xlabel("DATA_X")
ax.set_ylabel("DATA_Y")
ax.legend()
ax.set_title("Sample Data Plot")
```
